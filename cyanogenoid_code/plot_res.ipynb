```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
name = 'duke_testing' # change appropriately
    
with open(name + '_results.pkl', 'rb') as f:
    train_acc, train_loss, val_acc = pickle.load(f)
    
train_acc = np.concatenate((train_acc, train_acc))
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(7,6))
axes[0].plot([sum(train_loss[epoch])/len(train_loss[epoch]) for epoch in range(len(train_loss))], label='Training Loss')
axes[1].plot([sum(train_acc[epoch])/len(train_acc[epoch]) for epoch in range(len(train_acc))], label='Training Accuracy')
axes[2].plot([sum(val_acc[epoch])/len(val_acc[epoch]) for epoch in range(len(val_acc))], label='Validation Accuracy')

axes[0].set_xticks(np.arange(0, len(train_loss), 1.0))
axes[1].set_xticks(np.arange(0, len(train_acc), 1.0))
axes[2].set_xticks(np.arange(0, len(val_acc), 1.0))

axes[2].set_xlabel('Epoch')

axes[0].legend()
axes[1].legend()
axes[2].legend()
fig.suptitle('Results of ' + name)
```
